## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

**Exibir o número de dados faltantes para cada coluna do dataframe**

```python
def exibir_nans(dataset):
    n_linhas = len(dataset) # número total de linhas 
    colunas_nans = dataset.isnull().sum() # número de NaNs por coluna

    # número absoluto de NaNs e percentual equivalente
    print("-- DADOS FALTANTES --\n")
    for coluna, qtde_nans in colunas_nans.items():
        print(f'{coluna}: {qtde_nans} [{(qtde_nans / n_linhas)*100:0.2f}%]')

exibir_nans(invest_cadastro)
```

**Exibir gráfico com a porcentagem de dados faltantes nas colunas**

```python
import matplotlib.pyplot as plt

def plotar_grafico_nans(dataset):
    n_linhas = len(dataset) # número total de linhas 
    colunas_nans = dataset.isnull().sum() # número de NaNs por coluna
    
    labels = colunas_nans.index.to_list()
    values = colunas_nans.values / n_linhas * 100
    color = '#2ca25f'

    fig_nulls, ax_nulls = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
    ax_nulls.bar(
        [i for i in range(0, len(labels))],
        values,
        color=color,
    )

    ax_nulls.set_xticks([i for i in range(0, len(labels))])
    ax_nulls.set_xticklabels(labels, fontsize=7, rotation=90)
    ax_nulls.set_ylabel('Dados faltantes')
    ax_nulls.set_xlabel('Colunas')
    ax_nulls.set_title('Número percentual de dados faltantes por coluna')

plotar_grafico_nans(invest_cadastro)
```

**Exibir somente colunas com dados faltantes**

```python
def retornar_colunas_nans(dataset):
    colunas_nans = dataset.isnull().sum()
    values = colunas_nans.values
    lista_colunas_nans_true = colunas_nans[(values != 0)].index.tolist()
    return lista_colunas_nans_true

print(retornar_colunas_nans(invest_cadastro))
```

**Análise dos dados dos ativos**

```python
import seaborn as sns

lista_ativos = [
    'Ativo01',
    'Ativo02',
    'Ativo03',
    'Ativo04',
    'Ativo05',
    'Ativo06',
    'Ativo07'
]

sns.pairplot(invest_cadastro[lista_ativos])
# Não todos os ativos seguem uma distribuição Gaussiana, 
# e também não é evidente qualquer correlação entre os ativos.
```

**Preencher os dados faltantes**

```python
# https://towardsdatascience.com/handling-missing-data-like-a-pro-part-3-model-based-multiple-imputation-methods-bdfe85f93087
# https://towardsdatascience.com/whats-the-best-way-to-handle-nan-values-62d50f738fc

def fix_nan_datasets(original_dataset, columns):
    """Fix NaNs in `columns` from `original_dataset`.

    Parameters
    ----------
    original_dataset : pd.DataFrame
        Dataset
    columns : list
        Columns in `original_dataset` to use and fix NaNs values.

    Returns
    -------
    TODO
    """
    
    # dataset somente com as colunas que devem ser preenchidas
    df_to_imputed = original_dataset[columns]
    
    # criação do ketnel 
    kernel = mf.ImputationKernel(
        df_to_imputed, datasets=1, save_all_iterations=True, random_state=1
    )
    
    # execução do algoritmo mice por 10 iterações do 
    kernel.mice(10)
    
    # geração do dataset sem os valores NaNs 
    dataset_recovered = kernel.complete_data()
    return dataset_recovered


lista_ativos = [
    'Ativo01',
    'Ativo02',
    'Ativo03',
    'Ativo04',
    'Ativo05',
    'Ativo06',
    'Ativo07'
]

# Preenchimento dos dados faltantes dos ativos 
dataset_recovered = fix_nan_datasets(invest_cadastro, lista_ativos)

# Atribuir ao dataset original as colunas dos ativos depois do tratamento de dados faltantes
for column in dataset_recovered.columns.to_list():
    invest_cadastro[column] = dataset_recovered[column]


lista_notas = ['Nota1', 'Nota2']

# Preenchimento dos dados faltantes das notas
dataset_recovered = fix_nan_datasets(invest_cadastro, lista_notas)

# Atribuir ao dataset original as colunas dos ativos depois do tratamento de dados faltantes
for column in dataset_recovered.columns.to_list():
    invest_cadastro[column] = dataset_recovered[column]
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
invest_cadastro['Total'] = invest_cadastro[lista_ativos].sum(axis=1)

invest_cadastro['Media'] = invest_cadastro[lista_notas].mean(axis=1)

invest_cadastro['Dominio'] = invest_cadastro['Email'].apply(
    lambda x: x.split('@')[0]
)

# Coluna "Renda"
invest_cadastro = invest_cadastro.rename(columns={'Salario' : 'Renda'})
invest_cadastro['Renda'] = invest_cadastro['Renda'].apply(
    lambda x: x.replace('$', '')
)
invest_cadastro['Renda'] = invest_cadastro['Renda'].astype('float')

# Coluna "Sexo"
mapper = {'Masc': 'Masculino', 'Fem': 'Feminino'}
invest_cadastro['Sexo'] = invest_cadastro['Sexo'].map(mapper)

invest_cadastro['Media'] = invest_cadastro['Media'].astype('int')

def definir_nps(media):
    if media >= 80:
        return 'promotor'
    elif 50 <= media < 80:
        return 'neutro'
    return 'detrator'

invest_cadastro['NPS'] = invest_cadastro['Media'].apply(definir_nps)
```
