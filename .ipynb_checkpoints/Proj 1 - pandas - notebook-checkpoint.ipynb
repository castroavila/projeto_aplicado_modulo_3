## 1 - Carregar os arquivos cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
import pandas as pd
import datetime as dt
import miceforest as mf
import numpy as np

# leitura dos arquivos de investimento e de cadastro
arquivo_invest1 = pd.read_csv('datasets/investimento_parte1.csv', sep=',')
arquivo_invest2 = pd.read_csv('datasets/investimento_parte2.csv', sep=',')
cadastro = pd.read_csv('datasets/cadastro.csv', sep=',')

print('Arquivo investimentos 1')
print(arquivo_invest1.info()) # 12 colunas, 20000 linhas
print('-'*50)
print('-'*50)
print('-'*50)
print('Arquivo investimentos 2')
print(arquivo_invest2.info()) # 12 colunas, 24052 linhas
print('-'*50)
print('-'*50)
print('-'*50)
print('Arquivo cadastro') # 9 colunas, 38754 linhas
print(cadastro.info()) 
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
investimentos = pd.concat([arquivo_invest1, arquivo_invest2])
print(investimentos.info()) # 44052 linhas
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
investimentos = investimentos.drop_duplicates()
print(investimentos.shape) # quantidade de linhas x colunas 
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
invest_cadastro = investimentos.merge(cadastro, on='id')
print(invest_cadastro.shape) # quantidade de linhas x colunas
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
invest_cadastro = invest_cadastro.sort_values(by='id').set_index('id')
print(invest_cadastro)
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

**Exibir o número de dados faltantes para cada coluna do dataframe**

```python
n_linhas = len(invest_cadastro) # número total de linhas 
colunas_nans = invest_cadastro.isnull().sum() # número de NaNs por coluna

# número absoluto de NaNs e percentual equivalente
print("DADOS FALTANTES: ")
for coluna, values in colunas_nans.items():
    print(f'{coluna}: {values} [{(values/n_linhas)*100:0.2f}%]')
```

**Análise dos dados dos ativos**

```python
import seaborn as sns

lista_ativos = [
    'Ativo01',
    'Ativo02',
    'Ativo03',
    'Ativo04',
    'Ativo05',
    'Ativo06',
    'Ativo07'
]

lista_notas = ['Nota1', 'Nota2']
sns.pairplot(invest_cadastro[lista_ativos])
# Não todos os ativos seguem uma distribuição Gaussiana, e também não é evidente qualquer
# correlação entre os ativos.
```

**Preencher os dados faltantes**

```python
# https://towardsdatascience.com/handling-missing-data-like-a-pro-part-3-model-based-multiple-imputation-methods-bdfe85f93087
# https://towardsdatascience.com/whats-the-best-way-to-handle-nan-values-62d50f738fc

def fix_nan_datasets(original_dataset, columns):
    """Fix NaNs in `columns` from `original_dataset`.

    Parameters
    ----------
    original_dataset : pd.DataFrame
        Dataset
    columns : list
        Columns in `original_dataset` to use and fix NaNs values.

    Returns
    -------
    TODO
    """
    
    # dataset somente com as colunas que devem ser preenchidas
    df_to_imputed = original_dataset[columns]
    
    # criação do ketnel 
    kernel = mf.ImputationKernel(
        df_to_imputed, datasets=1, save_all_iterations=True, random_state=1
    )
    
    # execução do algoritmo mice por 10 iterações do 
    kernel.mice(10)
    
    # geração do dataset sem os valores NaNs 
    dataset_recovered = kernel.complete_data()

    return dataset_recovered


lista_ativos = [
    'Ativo01',
    'Ativo02',
    'Ativo03',
    'Ativo04',
    'Ativo05',
    'Ativo06'
]

# Preenchimento dos dados faltantes dos ativos 
dataset_recovered = fix_nan_datasets(invest_cadastro, lista_ativos)

# Atribuir ao dataset original as colunas dos ativos depois do tratamento de dados faltantes
for column in dataset_recovered.columns.to_list():
    invest_cadastro[column] = dataset_recovered[column]


lista_notas = ['Nota1', 'Nota2']
invest_cadastro_2 = invest_cadastro.dropna()
print(invest_cadastro_2.shape)

# Preenchimento dos dados faltantes das notas
dataset_recovered = fix_nan_datasets(invest_cadastro, lista_notas)

# Atribuir ao dataset original as colunas dos ativos depois do tratamento de dados faltantes
for column in dataset_recovered.columns.to_list():
    invest_cadastro[column] = dataset_recovered[column]
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
invest_cadastro['Total'] = invest_cadastro[lista_ativos].sum(axis=1)

invest_cadastro['Media'] = invest_cadastro[lista_notas].mean(axis=1)

invest_cadastro['Dominio'] = invest_cadastro['Email'].apply(
    lambda x: x.split('@')[0]
)

invest_cadastro = invest_cadastro.rename(columns={'Salario': 'Renda'})
invest_cadastro['Renda'] = invest_cadastro['Renda'].apply(
    lambda x: x.replace('$', '')
)
invest_cadastro['Renda'] = invest_cadastro['Renda'].astype('float')

mapper = {'Masc': 'Masculino', 'Fem': 'Feminino'}
invest_cadastro['Sexo'] = invest_cadastro['Sexo'].map(mapper)

invest_cadastro['Media'] = invest_cadastro['Media'].astype('int')

invest_cadastro['NPS'] = invest_cadastro['Media'].apply(
    lambda x: 'promotor'
    if x >= 80
    else 'neutro'
    if x >= 50 and x < 80
    else 'detrator'
)
```

## 8 - Calcule os seguintes dados
- Calcule o total investido em cada coluna Ativo (tente usar o apply)
- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo
- Qual a idade do investidor mais velho?
- Qual a mediana da coluna Média?

```python
print('Calcule o total investido em cada coluna Ativo (tente usar o apply)')
total_invest_ativo = invest_cadastro[lista_ativos].apply(np.sum, axis=0)
print(total_invest_ativo)
print('-' * 50)

print('Qual o menor valor investido em cada ativo?')
menor_valor_invest_ativo = invest_cadastro[lista_ativos].apply(np.min, axis=0)
print(menor_valor_invest_ativo)
print('-' * 50)

print('Mostre a média e desvio padrão de cada ativo')
print('Media')
media = invest_cadastro[lista_ativos].apply(np.mean, axis=0)
print(media)
print('-' * 30)
print('Desvio padrao')
std = invest_cadastro[lista_ativos].apply(np.std, axis=0)
print(std)
print('-' * 50)

print('Qual a idade do investidor mais velho?')
max_idade = invest_cadastro['Idade'].max()
print(max_idade)
print('-' * 50)

print('Qual a mediana da coluna Média?')
mediana = invest_cadastro['Media'].median()
print(mediana)
```

```python
## 9 - Responder as perguntas abaixo

- Quantos investidores são promotores?
- Quantos investidores são do sexo feminino?
- Quantos investidores investiram mais de 16000 no Ativo04?
- Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?
- Quantos investidores do sexo feminino tem mais de 25 anos e investiu menos de 10000? (bonus: por que esse dado seria importante para a estratégia da nossa empresa?

- Quantos investidores tem em cada estado (dica: use groupby)
- Qual é a soma do total investido por cada estado?
- Como os investidores se distribuem em promotores, detratores e neutros?
- Qual é a média do total investido por promotores, detratores e neutros?
```

```python
print('Quantos investidores são promotores?')
n_promotores = (invest_cadastro['NPS'] == 'promotor').sum()
print(n_promotores)
print('-' * 50)

print('Quantos investidores são do sexo feminino?')
n_feminino = (invest_cadastro['Sexo'] == 'Feminino').sum()
print(n_feminino)
print('-' * 50)

print('Quantos investidores investiram mais de 16000 no Ativo04?')
n_ativo04_acima_16000 = (invest_cadastro['Ativo04'] > 16000).sum()
print(n_ativo04_acima_16000)
print('-' * 50)

print(
    'Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?'
)
n_ativo04_ativo5 = (
    (invest_cadastro['Ativo04'] > 16000) & (invest_cadastro['Ativo05'] < 5000)
).sum()
print(n_ativo04_ativo5)
print('-' * 50)

print(
    'Quantos investidores do sexo feminino tem mais de 25 anos e investiu '
    + 'menos de 10000? (bonus: por que esse dado seria importante para a '
    + 'estratégia da nossa empresa?'
)

# Aqui assume-se total investido nos seis ativos.
n_investidoras = invest_cadastro[
    (invest_cadastro['Sexo'] == 'Feminino')
    & (invest_cadastro['Total'] < 10000)
    & (invest_cadastro['Idade'] > 25)
]
print(n_investidoras.size)
print('-' * 50)

print('Quantos investidores tem em cada estado (dica: use groupby)')
n_inv_estados = invest_cadastro.groupby('Estado')['Estado'].count()
print(n_inv_estados)
print('-' * 50)

print('Qual é a soma do total investido por cada estado?')
# Total investido = Ativo01 + Ativo02 + Ativo03 + ....
total_inves_estado = invest_cadastro.groupby('Estado')['Total'].sum()
print(total_inves_estado)
print('-' * 50)

print(
    'Como os investidores se distribuem em promotores, detratores e neutros?'
)
distr_investidores = invest_cadastro.groupby('NPS')['NPS'].count()
print(distr_investidores)
print('-' * 50)

print(
    'Qual é a média do total investido por promotores, detratores e neutros?'
)
media_investimento = invest_cadastro.groupby('NPS')['Total'].mean()
print(media_investimento)
```
